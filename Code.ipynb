```python
import csv
import pandas as pd
 

# Reading a file as a pd DataFrame
dataframe = pd.read_csv("denovo-db.ssc-samples.variants.v.1.6.1.tsv", sep="\t", skiprows=1, low_memory=False)  # the columns are loaded by default as the first row
df_non = pd.read_csv("denovo-db.non-ssc-samples.variants.v.1.6.1.tsv", sep="\t", skiprows=1, low_memory=False)
  
#dataframe = pd.concat([dataframe, df_non], axis=0)
```

```python
# filtering for missense varients

condition = dataframe["FunctionClass"] == "missense"
new_dataframe_ssc = dataframe[condition]

condition = df_non["FunctionClass"] == "missense"
new_dataframe_non_ssc = df_non[condition]
```

```python

# Reading a file of n lines into a list
with open("ASD_1_genes.txt") as gene_file:
  ASD = gene_file.read().splitlines()
```

```python
final_dataframe_ssc = new_dataframe_ssc[new_dataframe_ssc['Gene'].isin(ASD)]

# filtering for specifically autism or control samples
final_dataframe_ssc = final_dataframe_ssc[(final_dataframe_ssc['PrimaryPhenotype'] == "autism") | (final_dataframe_ssc['PrimaryPhenotype'] == "control")]

final_dataframe_non_ssc = new_dataframe_non_ssc[new_dataframe_non_ssc['Gene'].isin(ASD)]

# filtering for specifically autism or control samples
final_dataframe_non_ssc = final_dataframe_non_ssc[(final_dataframe_non_ssc['PrimaryPhenotype'] == "autism") | (final_dataframe_non_ssc['PrimaryPhenotype'] == "control")]

```

```python

# Selecting only the columns we're interested in
df_ssc = final_dataframe_ssc[['PrimaryPhenotype', 'Chr', 'Position', 'Variant', 'Gene', 'FunctionClass','rsID']]


df_non_ssc = final_dataframe_non_ssc[['PrimaryPhenotype', 'Chr', 'Position', 'Variant', 'Gene', 'FunctionClass','rsID']]

df_non_ssc
```

```python
dataframe = pd.concat([df_ssc, df_non_ssc], axis=0)
dataframe
```
